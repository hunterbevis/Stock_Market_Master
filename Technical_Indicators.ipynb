```
# gather data and plot BBands
#    matype :  Moving average type. By default, matype=0.
#        Integers 0 - 8 are accepted (check  down the mappings) or the string
#        containing the math type can also be used.

#    * 0 = Simple Moving Average (SMA),
#    * 1 = Exponential Moving Average (EMA),
#    * 2 = Weighted Moving Average (WMA),
#    * 3 = Double Exponential Moving Average (DEMA),
#    * 4 = Triple Exponential Moving Average (TEMA),
#    * 5 = Triangular Moving Average (TRIMA),
#    * 6 = T3 Moving Average,
#    * 7 = Kaufman Adaptive Moving Average (KAMA),
#    * 8 = MESA Adaptive Moving Average (MAMA)

# Get BBands data
try:
  data1, meta_data1 = ti.get_bbands(symbol=stockTicker, interval='60min', time_period=60)
except:
  if (key_picker == 3):
    key_picker = 1
    data1, meta_data1 = ti.get_bbands(symbol=stockTicker, interval='60min', time_period=60)
  else:
    key_picker +=1
    data1, meta_data1 = ti.get_bbands(symbol=stockTicker, interval='60min', time_period=60)

#Plot

fig1 = px.line(title= "BB-bands indicator for {}".format(stockTicker))
for i in data1.columns[0:]:
  fig1.add_scatter(x = data1.index, y = data1[i], name= i)
fig1.update_layout(yaxis_title='Simple Moving Average (SMA)',
                   xaxis_title="Tight bands indicate low volatility (low price fluctuation). Bands far apart indicate period of high volatility (high price fluctuation)")

data1.describe()
```
