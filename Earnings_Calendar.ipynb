```python
apiKey = "d8955c763cf41e8593a8bd57288aa7d3"
fromDate = '2021-05-10'
toDate = '2021-06-10'

# fromDate = '2021-05-15'
# toDate = '2021-05-21'

# fromDate = '2021-05-22'
# toDate = '2021-05-31'
```

```python
earningsCalendar = requests.get(f'https://financialmodelingprep.com/api/v3/earning_calendar?from={fromDate}&to={toDate}&apikey={apiKey}')
EC = earningsCalendar.json()
```

```python
earnings_df = pd.json_normalize(EC)
```

```python
stockList = earnings_df['symbol'].tolist()
```

```python

timeCount = 0
industryList = []
descList = []
for stock in stockList:
  timeCount += 1
  if timeCount >= 295:
    print(timeCount)
    timeCount = 1
    time.sleep(61)
  CP = requests.get(f'https://financialmodelingprep.com/api/v3/profile/{stock}?apikey={apiKey}')
  cp_json = CP.json()
  try:
    industryList.append(cp_json[0]['industry'])
    descList.append(cp_json[0]['description'])
  except:
    industryList.append('null')
    descList.append('null')


# Build DF with new lists
earnings_df['Industry'] = industryList
earnings_df['Description'] = descList

print(earnings_df)
```

```python
# printing output
import xlsxwriter
import re

master_df = earnings_df.copy()
master_df.to_excel("Earnings_2021-05-10_2021-06-10.xlsx")
```
